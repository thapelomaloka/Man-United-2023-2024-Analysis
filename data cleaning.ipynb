## Clean the Player stats Table

```python
# Remove 'm' from the 'HT' column

PlayerStats["HT"] = PlayerStats["HT"].str.replace("m", "")
PlayerStats["WT"] = PlayerStats["WT"].str.replace("kg", "")
PlayerStats
```

```python
PlayerStats.to_csv("PlayerStatsClean.csv")
```
