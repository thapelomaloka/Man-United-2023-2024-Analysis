```python
# Importing necessary libraries
import pandas as pd
import numpy as np

# For data visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
standardStats = pd.read_csv('Data Sets\Raw Data\Standard Stats 2023-2024 Manchester United.csv')
```

```python
#View the 1st 5 rows

standardStats.head()
```

```python
#Function to drop columns
def drop_columns(dataframe, columns_to_drop):
    """
    Drops specified columns from the dataframe.
    
    Parameters:
    dataframe (pd.DataFrame): The DataFrame from which columns are to be dropped.
    columns_to_drop (list): List of column names to drop.
    
    Returns:
    pd.DataFrame: The DataFrame with specified columns dropped.
    """
    dataframe.drop(columns=columns_to_drop, inplace=True)
    return dataframe
```

```python
columns=['90s','PKatt','xG','npxG','xAG','npxG+xAG']

standardStats= drop_columns(standardStats,columns)

standardStats.head()
```

```python
standardStats.isnull().sum()
```

```python
# Fill NaN values with 0 becuase the players did not play any games
standardStats = standardStats.fillna(0)
```

```python
# Convert specific columns from float to int
columns_to_convert = ['GOALS', 'ASISTS', 'GOALS & ASSISTS', 'NON-PENALTY GOALS', 'PENALTY KICK', 'YELLOW CARDS', 'RED CARDS']

# Use a loop to convert each column
for column in columns_to_convert:

    standardStats[column] = standardStats[column].astype(int)
    
standardStats.head()
```
